# Chapter2. 데이터 핸들링

```python
import pandas as pd
import numpy as np

students = pd.DataFrame({'이름':['영희','영숙','정숙','영희','숙자'],
                        '국어' : [65,43,26,78,99],
                        '수학' : [None, 53,98,76,None],
                        '영어' : [None,34,79,60,66],
                        '과학' : [None, 38,87,99,69]})
students
```

## 데이터탐색

- np.select(조건목록, 선택목록, default = 디폴트값)

```python
import numpy as np
condition_list = [(students['국어'] >= 90),
                  (students['국어'] >= 80) & (students['국어'] < 90),
                  (students['국어'] >= 70) & (students['국어'] <= 80)]
choice_list = ['A','B','C']
students['점수'] = np.select(condition_list, choice_list, default = 'E')
students
```

- 결측값 제거 (dropna 사용)

```python
# 결측값이 있는 행 제거
students.dropna()
```

```python
# 결측값이 1개 초과인 경우 해당 행을 제거
students.dropna(thresh = 1)
```

```python
# 결측값 처리 - 이전 열의 값과 동일한 값으로 대체
students.fillna(method = 'pad')
```
