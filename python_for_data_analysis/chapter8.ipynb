# Chapter8: 데이터 준비하기: 조인, 병합, 변형

```python
import pandas as pd
df1 =pd.DataFrame([1,4,5,6])
df2 = pd.DataFrame([6,3,7,1])
df3 = pd.DataFrame([6,8,2,1])
pd.concat([df1,df2,df3])
```

```python
# Stack형태로 추출
test_stacked = pd.concat([df1,df2,df3], keys = ['one', 'two','three'])
test_stacked
```

```python
test_stacked.unstack()
```

```python
# Stack형태로 추출
test_stacked = pd.concat([df1,df2,df3], axis = 1, keys = ['one', 'two','three'], names = ['first','second'], ignore_index=False)
test_stacked
```

```python
df = pd.DataFrame({'Key':['foo','bar','zoo'],
                    'A':[1,2,3],
                    'B':[4,5,6],
                    'C':[7,8,9]})
df
```

```python
melted = pd.melt(df, ['Key'])
melted
```

```python
melted.pivot('Key','variable','value')
```

```python
melted = pd.melt(df, id_vars = ['Key'], value_vars = ['A','B'])
melted
```
